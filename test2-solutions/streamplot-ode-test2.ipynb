```python
import matplotlib.pyplot as plt
%matplotlib inline 
%config InlineBackend.figure_format = 'svg'
import numpy as np
import cmath
```

```python
plt.figure(figsize=(6,6))
plt.xlabel("$x$")
plt.ylabel("$y$")
y, x = np.mgrid[-5:5:5000j, -5:5:5000j] 

plt.streamplot(x, y, 3 * x - 2 * y, 2 * x - 2 * y)
plt.axis('square')
plt.axis([-5, 5, -5, 5])
plt.arrow(0,0,1,2, color='r', head_width=0.15, width = 0.03, length_includes_head='True')
plt.arrow(0,0,2,1, color='r', head_width=0.15, width = 0.03, length_includes_head='True')
plt.grid()
plt.savefig("1a1_1.pdf")
plt.figure(figsize=(6,6))
plt.xlabel("$x$")
plt.ylabel("$y$")
y, x = np.mgrid[-4:4:50j, -4:4:50j] 
plt.streamplot(x, y, -x, 2*y)
plt.axis('square')
plt.axis([-4, 4, -4, 4])
plt.arrow(0,0,1,0, color='r', head_width=0.15, width = 0.03, length_includes_head='True')
plt.arrow(0,0,0,1, color='r', head_width=0.15, width = 0.03, length_includes_head='True')
plt.grid()
plt.savefig("1a1_0.pdf")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.xlabel("$x$")
plt.ylabel("$y$")
y, x = np.mgrid[-5:5:5000j, -5:5:50j] 

t = np.arange(1, 100, 0.1)
for i in range(1, 60, 6):
    x = i/10 * np.sin(1 * t)
    y = i/10 * np.cos(1 * t)
    plt.plot(x, y, 'b')
plt.axis('square')
plt.axis([-4, 4, -4, 4])
plt.grid()
plt.savefig("3a1_0.pdf")
plt.show()

plt.figure(figsize=(6,6))
plt.xlabel("$x$")
plt.ylabel("$y$")
y, x = np.mgrid[-5:5:5000j, -5:5:50j] 

plt.streamplot(x, y, x - y, 5 * x - y)
plt.axis('square')
plt.axis([-4, 4, -4, 4])
plt.grid()
plt.savefig("3a1_1.pdf")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.xlabel("$r\alpha_1$")
plt.ylabel("$r\alpha_2$")
y, x = np.mgrid[-5:5:50j, -5:5:50j] 

plt.streamplot(x, y, 2*x+y, 2*y)
plt.axis('square')
plt.axis([-4, 4, -4, 4])
plt.grid()
plt.savefig("1b1_0.pdf")
plt.show()



plt.figure(figsize=(6,6))
plt.xlabel("$x$")
plt.ylabel("$y$")
y, x = np.mgrid[-5:5:50j, -5:5:50j] 

plt.streamplot(x, y, 3*x + y, - x + y)
plt.arrow(0,0,-1,1, color='r', head_width=0.15, width = 0.05, length_includes_head='True')
plt.arrow(0,0,1,0, color='r', head_width=0.15, width = 0.05, length_includes_head='True')
plt.axis('square')
plt.axis([-4, 4, -4, 4])
plt.grid()
plt.savefig("1b1_1.pdf")
plt.show()
```
