```python
import matplotlib.pyplot as plt
%matplotlib inline 
%config InlineBackend.figure_format = 'svg'
import numpy as np
import cmath
```

### This function builds a phase portrait in original and canonical coordinates and calculates eigenvalues and eigenvectors

```python
def streamplot(a, b, c, d):
    plt.figure(figsize=(6,6))
    plt.xlabel("$x$")
    plt.ylabel("$y$")
    y, x = np.mgrid[-5:5:5000j, -5:5:5000j] 
    mat = np.array([[a, b], [c, d]])
    plt.streamplot(x, y, a * x + b * y, c * x + d * y)
    eigval, eigvec = np.linalg.eig(mat)
    print('eigenvalues:', eigval, 'eigenvectors:', eigvec)
    if eigvec[0,0].imag == 0 and eigvec[0,1].imag == 0 and eigvec[1,1].imag == 0 and eigvec[1,0].imag == 0:
        plt.arrow(0,0,eigvec[0,0],eigvec[1,0], color='r', head_width=0.15, width = 0.03, length_includes_head='True')
        plt.arrow(0,0,eigvec[0,1],eigvec[1,1], color='r', head_width=0.15, width = 0.03, length_includes_head='True')
    plt.grid()
    plt.figure(figsize=(6,6))
    plt.xlabel("$x$")
    plt.ylabel("$y$")
    if eigvec[0,0].imag == 0 and eigvec[0,1].imag == 0 and eigvec[1,1].imag == 0 and eigvec[1,0].imag == 0:
        plt.streamplot(x, y, eigval[0]*x, eigval[1]*y)
        plt.axis([-4, 4, -4, 4])
        plt.arrow(0,0,1,0, color='r', head_width=0.15, width = 0.03, length_includes_head='True')
        plt.arrow(0,0,0,1, color='r', head_width=0.15, width = 0.03, length_includes_head='True')
    else:
        plt.streamplot(x, y, eigval[0].real*x+eigval[0].imag*y, eigval[1].imag*x+eigval[1].real*y)
        plt.axis([-4, 4, -4, 4])
    plt.grid()
    plt.show()
```

### Some examples: test-2 solutions (2021/2022)

```python
streamplot(5, -2, 4, -1)
```
